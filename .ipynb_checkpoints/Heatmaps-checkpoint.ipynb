```python
def growthrate_temp(T, Tmin = 3.99, Tmax = 43.7, b3 = .0410, c3 = .161):
    '''
    This function takes an input of a single temperature and outputs a growthrate 
    value at that temperature based on the Ratkowsky 3 growthrate vs. temperature 
    model and the parameters described in Zwiettering et al. 
    '''    
    muM = np.power(b3*(T-Tmin), 2)*(1-np.power(np.e, (c3*(T-Tmax))))
    
    if (muM < 0):
        muM = 0
    return muM

def growthrate_conc(allele, temp, concentration, c = -.6824968):
    '''
   Takes as input an MIC, g_drugless, and concentration, and fitting constant, and outputs
   a growthrate at that drug concentration as defined by the logistic curve described by 
   Ogbunufor et al. 
   '''
    g_drugless = growthrate_temp(temp)
    MIC = tca.MIC_fromT(allele, temp)
    growthrate = g_drugless/(1+19*np.e**((np.log10(MIC)-np.log10(concentration))/c))
    return growthrate

def conc_curve(allele, temp, concentrations, c = -.6824968):
    '''
    Takes as inputs a single MIC, drugless growthrate, array of concentration values, and 
    a logistic fitting constant from Ogbunufor et al. and outputs an array of growthrate values as 
    a function of concentration.
    '''
    
    g_drugless = growthrate_temp(temp)
    MIC = tca.MIC_fromT(allele, temp)
    
    growthrates = []
    concs_growths = []
    for i in range(len(concentrations)):
        growthrate = growthrate_conc(MIC, g_drugless, concentrations[i], c)
        conc_growth = [(np.log10(concentrations[i]), growthrate)]
        concs_growths.append(conc_growth)
        
    concs_growths = np.array(concs_growths)
    concs_growths = concs_growths[:, 0, :] #just slices out the 2nd dimension of the concs_growths array -- somehow I have an extra one 
    return concs_growths
```

```python
def interpolation_function(allele):

    concs = np.power(10., np.arange(-10, 10, 1))
    temps_exp = np.array([20, 25, 30, 35, 37, 41])
    
    T, C = np.meshgrid(temps_exp, concs)

    growthrate_grid = np.ones((T.shape[0], T.shape[1]))
    
    for i in range(growthrate_grid.shape[0]):
        for j in range(growthrate_grid.shape[1]):
            growthrate_grid[i, j] = growthrate_conc(allele = allele, temp = T[i, j], concentration = C[i, j], c = -.6824968)

    growthrate_function = inp.interp2d(temps_exp, concs, growthrate_grid, kind = "cubic")

    #closer_temps = np.arange(18, 43, .1)
    #closer_concs = np.power(10., np.arange(-10, 10, .1))
    #TT, CC = np.meshgrid(closer_temps, closer_concs)

    #growthrate_tightgrid = growthrate_function(closer_temps, closer_concs)
    
    return growthrate_function
    
```

```python
interpolation_functions = []
for i in range(32):
    interpolation_functions.append(interpolation_function(i))
interpolation_functions = np.array(interpolation_functions)

print(interpolation_functions[27](27, 5))
    
```
