```python
from pylab import *
from matplotlib import rc
import matplotlib.gridspec as gridspec
import numpy as np

from matplotlib.font_manager import FontProperties
import matplotlib.colors as colors

import matplotlib.cm as cm
from matplotlib import patches

from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
%matplotlib notebook
clf()

xoff=0.2
yoff=0.2
width=0.85-xoff
height=0.85-yoff

dashes=[3,2]
labelx=-0.11

##################
fig_width_pt = 246.0
inches_per_pt = 1.0/72.27
golden_mean = (sqrt(5)-1.0)/2.0
fig_width = fig_width_pt*inches_per_pt
fig_height = fig_width_pt*inches_per_pt*golden_mean
fig_size =  [1.5*fig_width,2.5*fig_height]
params = {'backend': 'pdf',
  'axes.labelsize': 9,
  #'text.fontsize': 9,
  'xtick.labelsize': 9,
  'ytick.labelsize': 9,
  'text.usetex': False,
  'figure.figsize': fig_size,
  'figure.subplot.left': xoff,
  'figure.subplot.bottom': yoff,
  'figure.subplot.right': xoff+width,
  'figure.subplot.top': yoff+height,
                  'mathtext.fontset': 'stixsans'}

rcParams.update(params)
rc('font',**{'family':'sans-serif','sans-serif':['Tex Gyre Heros']})
        
figure(1)

gs1=gridspec.GridSpec(1,1)
gs1.update(hspace=0.0,wspace=0)

cols={"teal": '#469990',"cyan" : '#42d4f4',"green" : '#3cb44b', "red" : '#e6194B', "yellow" : '#ffe119', "blue" : '#4363d8', "orange" : '#f58231', "pink" : '#fabebe', "lavender" : '#e6beff', "maroon" : '#800000', "navy" : '#000075', "grey" : '#a9a9a9', "black" : '#000000'}
ylpos=-0.12


############################################3
        
subplot(gs1[0,0])

A2=np.array(np.loadtxt('klcdA2.txt',delimiter='\t'))
A5=np.array(np.loadtxt('klcdA5.txt',delimiter='\t'))
A10=np.array(np.loadtxt('klcdA10.txt',delimiter='\t'))

A01=np.array(np.loadtxt('klcdA01.txt',delimiter='\t'))
A02=np.array(np.loadtxt('klcdA02.txt',delimiter='\t'))
A05=np.array(np.loadtxt('klcdA05.txt',delimiter='\t'))

Bm0=np.array(np.loadtxt('klcd.txt',delimiter='\t'))
Bm0no=np.array(np.loadtxt('klnocd.txt',delimiter='\t'))

#plot([e[0] for e in Bm0],[e[1] for e in Bm0],lw=1.25,color='k')
lno,=plot([e[0] for e in Bm0no],[e[1] for e in Bm0no],lw=1.25,color='k')
lno.set_dashes([2,2])

l01,=plot([e[0] for e in A01],[e[1] for e in A01],lw=1.25,color=cols["red"])
l02,=plot([e[0] for e in A02],[e[1] for e in A02],lw=1.25,color=cols["orange"])
l05,=plot([e[0] for e in A05],[e[1] for e in A05],lw=1.25,color=cols["green"])

l2,=plot([e[0] for e in A2],[e[1] for e in A2],lw=1.25,color='c')
l5,=plot([e[0] for e in A5],[e[1] for e in A5],lw=1.25,color=cols["blue"])
l10,=plot([e[0] for e in A10],[e[1] for e in A10],lw=1.25,color='Purple')

#gca().set_yscale('log')
setp(gca(),xlim=[5,300],ylim=[1e-4,30])
xlabel(r'Time [generations]')
ylabel(r'$D_{\rm KL}(\rho\,||\,p)$ [bits]')
#gca().yaxis.set_label_coords(ylpos,0.5)

leg=legend([lno,l01,l02,l05,l2,l5,l10],[r'no CD',r'$c = 0.1$',r'$c = 0.2$',r'$c = 0.5$',r'$c = 2$',r'$c = 5$',r'$c = 10$'],handlelength=2.5,loc=(0.68,0.53),numpoints=3,prop = FontProperties(size=9),ncol=1,labelspacing=0.35,handletextpad=0.5)
leg.draw_frame(0)


############################################3

#figtext(0.22,0.82,'A',fontsize=12,fontweight='bold')
#figtext(0.22,0.57,'B',fontsize=12,fontweight='bold')

show()
```
