```python
from pylab import *
from matplotlib import rc
import matplotlib.gridspec as gridspec
import numpy as np

from matplotlib.font_manager import FontProperties
import matplotlib.colors as colors

import matplotlib.cm as cm
from matplotlib import patches

from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
%matplotlib inline
clf()

xoff=0.2
yoff=0.2
width=0.85-xoff
height=0.85-yoff

dashes=[3,2]
labelx=-0.11

##################
fig_width_pt = 246.0
inches_per_pt = 1.0/72.27
golden_mean = (sqrt(5)-1.0)/2.0
fig_width = fig_width_pt*inches_per_pt
fig_height = fig_width_pt*inches_per_pt*golden_mean
fig_size =  [3.5*fig_width,3.5*fig_height]
params = {'backend': 'pdf',
  'axes.labelsize': 9,
  #'text.fontsize': 9,
  'xtick.labelsize': 9,
  'ytick.labelsize': 9,
  'text.usetex': False,
  'figure.figsize': fig_size,
  'figure.subplot.left': xoff,
  'figure.subplot.bottom': yoff,
  'figure.subplot.right': xoff+width,
  'figure.subplot.top': yoff+height,
                  'mathtext.fontset': 'stixsans'}

rcParams.update(params)
rc('font',**{'family':'sans-serif','sans-serif':['Tex Gyre Heros']})
        
figure(1)

gs1=gridspec.GridSpec(4,2)
gs1.update(hspace=0.0,wspace=0.25,bottom=0.3)

gs2=gridspec.GridSpec(1,2)
gs2.update(hspace=0.0,wspace=0.25,top=0.37)

cols={"teal": '#469990',"cyan" : '#42d4f4',"green" : '#3cb44b', "red" : '#e6194B', "yellow" : '#ffe119', "blue" : '#4363d8', "orange" : '#f58231', "pink" : '#fabebe', "lavender" : '#e6beff', "maroon" : '#800000', "navy" : '#000075', "grey" : '#a9a9a9', "black" : '#000000'}
ylpos=-0.12

############################################3
        
subplot(gs1[0,0])

hist=np.array(np.loadtxt('historig_50.txt',delimiter='\t'))
porig=np.array(np.loadtxt('porig_50.txt',delimiter='\t'))
peq=np.array(np.loadtxt('peq_50.txt',delimiter='\t'))

fill_between([e[0] for e in peq],0,[e[1] for e in peq],facecolor=cols["blue"],alpha=0.25)
leq,=plot([e[0] for e in peq],[e[1] for e in peq],lw=1.5,color=cols["blue"])

fill_between([e[0] for e in porig],0,[e[1] for e in porig],facecolor=cols["red"],alpha=0.25)
lorig,=plot([e[0] for e in porig],[e[1] for e in porig],lw=1.5,color=cols["red"])
hs,=plot([e[0] for e in hist],[e[1] for e in hist],'ko',mfc='None',mec=cols["red"],mew=1,ms=4.5)

setp(gca(),xlim=[0.25,1],ylim=[0,18.0],xticklabels=[])
ylabel(r'Probability')
gca().yaxis.set_label_coords(ylpos,0.5)

leg=legend([lorig,hs,leq],[r'$p\,(x_1,t\,)$ theory',r'$p\,(x_1,t\,)$ simulations',r'$\rho\,(x_1;\lambda(t))$'],handlelength=2.5,loc=(0.02,0.4),numpoints=3,prop = FontProperties(size=9),ncol=1,labelspacing=0.35,handletextpad=0.5)
leg.draw_frame(0)
text(0.91,14.5,r'$t=50$',fontsize=9)
title('Original protocol',fontsize=10)

############################################3
        
subplot(gs1[1,0])

hist=np.array(np.loadtxt('historig_100.txt',delimiter='\t'))
porig=np.array(np.loadtxt('porig_100.txt',delimiter='\t'))
peq=np.array(np.loadtxt('peq_100.txt',delimiter='\t'))

fill_between([e[0] for e in peq],0,[e[1] for e in peq],facecolor=cols["blue"],alpha=0.25)
plot([e[0] for e in peq],[e[1] for e in peq],lw=1.5,color=cols["blue"])

fill_between([e[0] for e in porig],0,[e[1] for e in porig],facecolor=cols["red"],alpha=0.25)
plot([e[0] for e in porig],[e[1] for e in porig],lw=1.5,color=cols["red"])
hs,=plot([e[0] for e in hist],[e[1] for e in hist],'ko',mfc='None',mec=cols["red"],mew=1,ms=4.5)

setp(gca(),xlim=[0.25,1],ylim=[0,18],xticklabels=[])
ylabel(r'Probability')
gca().yaxis.set_label_coords(ylpos,0.5)
text(0.91,14.5,r'$t=100$',fontsize=9)

############################################3
        
subplot(gs1[2,0])

hist=np.array(np.loadtxt('historig_150.txt',delimiter='\t'))
porig=np.array(np.loadtxt('porig_150.txt',delimiter='\t'))
peq=np.array(np.loadtxt('peq_150.txt',delimiter='\t'))

fill_between([e[0] for e in peq],0,[e[1] for e in peq],facecolor=cols["blue"],alpha=0.25)
plot([e[0] for e in peq],[e[1] for e in peq],lw=1.5,color=cols["blue"])

fill_between([e[0] for e in porig],0,[e[1] for e in porig],facecolor=cols["red"],alpha=0.25)
plot([e[0] for e in porig],[e[1] for e in porig],lw=1.5,color=cols["red"])
hs,=plot([e[0] for e in hist],[e[1] for e in hist],'ko',mfc='None',mec=cols["red"],mew=1,ms=4.5)

setp(gca(),xlim=[0.25,1],ylim=[0,18])
xlabel(r'Genotype fraction $x_1$')
ylabel(r'Probability')
gca().yaxis.set_label_coords(ylpos,0.5)
text(0.91,14.5,r'$t=150$',fontsize=9)

############################################3
        
subplot(gs1[0,1])

hist=np.array(np.loadtxt('histCD_50.txt',delimiter='\t'))
porig=np.array(np.loadtxt('pcd_50.txt',delimiter='\t'))
peq=np.array(np.loadtxt('peq_50.txt',delimiter='\t'))

fill_between([e[0] for e in peq],0,[e[1] for e in peq],facecolor=cols["blue"],alpha=0.25)
plot([e[0] for e in peq],[e[1] for e in peq],lw=1.5,color=cols["blue"])

fill_between([e[0] for e in porig],0,[e[1] for e in porig],facecolor=cols["red"],alpha=0.25)
plot([e[0] for e in porig],[e[1] for e in porig],lw=1.5,color=cols["red"])
hs,=plot([e[0] for e in hist],[e[1] for e in hist],'ko',mfc='None',mec=cols["red"],mew=1,ms=4.5)

setp(gca(),xlim=[0.25,1],ylim=[0,18.0],xticklabels=[])
ylabel(r'Probability')
gca().yaxis.set_label_coords(ylpos,0.5)
text(0.91,14.5,r'$t=50$',fontsize=9)
title('CD protocol',fontsize=10)


############################################3
        
subplot(gs1[1,1])

hist=np.array(np.loadtxt('histCD_100.txt',delimiter='\t'))
porig=np.array(np.loadtxt('pcd_100.txt',delimiter='\t'))
peq=np.array(np.loadtxt('peq_100.txt',delimiter='\t'))

fill_between([e[0] for e in peq],0,[e[1] for e in peq],facecolor=cols["blue"],alpha=0.25)
plot([e[0] for e in peq],[e[1] for e in peq],lw=1.5,color=cols["blue"])

fill_between([e[0] for e in porig],0,[e[1] for e in porig],facecolor=cols["red"],alpha=0.25)
plot([e[0] for e in porig],[e[1] for e in porig],lw=1.5,color=cols["red"])
hs,=plot([e[0] for e in hist],[e[1] for e in hist],'ko',mfc='None',mec=cols["red"],mew=1,ms=4.5)

setp(gca(),xlim=[0.25,1],ylim=[0,18],xticklabels=[])
ylabel(r'Probability')
gca().yaxis.set_label_coords(ylpos,0.5)
text(0.91,14.5,r'$t=100$',fontsize=9)


############################################3
        
subplot(gs1[2,1])

hist=np.array(np.loadtxt('histCD_150.txt',delimiter='\t'))
porig=np.array(np.loadtxt('pcd_150.txt',delimiter='\t'))
peq=np.array(np.loadtxt('peq_150.txt',delimiter='\t'))

fill_between([e[0] for e in peq],0,[e[1] for e in peq],facecolor=cols["blue"],alpha=0.25)
plot([e[0] for e in peq],[e[1] for e in peq],lw=1.5,color=cols["blue"])

fill_between([e[0] for e in porig],0,[e[1] for e in porig],facecolor=cols["red"],alpha=0.25)
plot([e[0] for e in porig],[e[1] for e in porig],lw=1.5,color=cols["red"])
hs,=plot([e[0] for e in hist],[e[1] for e in hist],'ko',mfc='None',mec=cols["red"],mew=1,ms=4.5)

setp(gca(),xlim=[0.25,1],ylim=[0,18])
xlabel(r'Genotype fraction $x_1$')
ylabel(r'Probability')
gca().yaxis.set_label_coords(ylpos,0.5)
text(0.91,14.5,r'$t=150$',fontsize=9)


############################################3

subplot(gs2[0,0])

sv=np.array(np.loadtxt('svtab.txt',delimiter='\t'))

plot([e[0] for e in sv],[e[1] for e in sv],lw=1.5,color=cols["navy"])
plot([e[0] for e in sv],[e[2] for e in sv],lw=1.5,color=cols["teal"])
plot([50,100,150],[0.043,0.043,0.043],'kv',color='Purple',ms=5)

setp(gca(),xlim=[0,300],ylim=[0,0.045])
xlabel(r'Time [generations]')
ylabel(r'Selection coefficient')
gca().yaxis.set_label_coords(ylpos,0.5)
text(130,0.031,r'CD',fontsize=9,color=cols["teal"])
text(130,0.01,r'original',fontsize=9,color=cols["navy"])


############################################3

subplot(gs2[0,1])

kl1=np.array(np.loadtxt('klnocd.txt',delimiter='\t'))
kl2=np.array(np.loadtxt('klcd.txt',delimiter='\t'))

plot([e[0] for e in kl1],[e[1] for e in kl1],lw=1.5,color=cols["navy"])
plot([e[0] for e in kl2],[e[1] for e in kl2],lw=1.5,color=cols["teal"])

gca().set_yscale('log')
setp(gca(),xlim=[0,300],ylim=[0.0001,100])
xlabel(r'Time [generations]')
ylabel(r'$D_{\rm KL}(\rho\,||\,p)$ [bits]')
gca().yaxis.set_label_coords(ylpos,0.5)

plot([50,100,150],[55,55,55],'kv',color='Purple',ms=5)


text(240,0.01,r'CD',fontsize=9,color=cols["teal"])
text(240,4,r'original',fontsize=9,color=cols["navy"])


############################################3

figtext(0.15,0.845,'A',fontsize=12,fontweight='bold')
figtext(0.15,0.38,'B',fontsize=12,fontweight='bold')
figtext(0.512,0.845,'C',fontsize=12,fontweight='bold')
figtext(0.512,0.38,'D',fontsize=12,fontweight='bold')


show()
```
